RNN con dos capas de 3 neuronas

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np
```

```python
numero_entradas = 2
numero_neuronas = 3
```

```python
x0 = tf.placeholder(tf.float32,[None, numero_entradas]) #entrada a capa 1
x1 = tf.placeholder(tf.float32,[None, numero_entradas]) #entrada a capa 2
```

```python
wx = tf.Variable(tf.random_normal(shape=[numero_entradas,numero_neuronas])) #pes 1
wy = tf.Variable(tf.random_normal(shape=[numero_neuronas,numero_neuronas])) #pes 2
```

```python
b = tf.Variable(tf.zeros([1,numero_neuronas])) #bias
```

```python
y0 = tf.tanh(tf.matmul(x0,wx) + b)#sortida de la primera capa
y1 = tf.tanh(tf.matmul(y0,wy) + tf.matmul(x1,wx) + b)#sortida de la segona capa
```

```python
lote_x0 = np.array([[0,1],[2,3],[4,5]]) #valors per testejar les entrades
lote_x1 = np.array([[2,4],[3,9],[4,1]])
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sesion:
    sesion.run(init)
    salida_y0, salida_y1 = sesion.run([y0,y1], feed_dict={x0:lote_x0, x1:lote_x1})
```
