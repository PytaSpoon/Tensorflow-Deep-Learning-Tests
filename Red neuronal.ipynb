# RED NEURONAL CON TENSORFLOW
### Z = W*X + B

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np
```

Iniciamos las variables random

```python
random_a = np.random.uniform(0,50,(4,4))
random_b = np.random.uniform(0,50,(4,1))
```

Creamos placeholders

```python
a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
```

```python
suma = a + b
mult = a * b
```

```python
with tf.Session() as sesion:
    res_suma = sesion.run(suma,feed_dict={a:random_a, b:random_b})
    res_mult = sesion.run(mult,feed_dict={a:random_a, b:random_b})
    print(res_suma)
    print('\n')
    print(res_mult)
```

## Red neuronal
Con 10 caracter√≠sticas y 4 neuronas.

```python
caracteristicas = 10
neuronas = 4
x = tf.placeholder(tf.float32,(None, caracteristicas))
W = tf.Variable(tf.random.normal([caracteristicas, neuronas]))
b = tf.Variable(tf.ones([neuronas]))
initializer = tf.global_variables_initializer()
```

```python
multip = tf.matmul(x,W)
z = tf.add(multip,b)
activation = tf.sigmoid(z)
x_values = np.random.random([1,caracteristicas])
```

```python
with tf.Session() as sesion:
    sesion.run(initializer)
    final_result = sesion.run(activation, feed_dict={x:x_values})
```

```python
print(final_result)
```
