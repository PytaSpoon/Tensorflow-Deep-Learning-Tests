```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
nodo1 = tf.constant(4)
nodo2 = tf.constant(6)
```

```python
nodo3 = nodo1 + nodo2
```

```python
with tf.Session() as sesion:
    result = sesion.run(nodo3)
    print(result)
```
