```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
mensaje1= tf.constant('Hola ')
mensaje2= tf.constant('Mundo')
```

```python
with tf.Session() as sesion:
    resultado = sesion.run(mensaje1 + mensaje2)
```

```python
resultado
```

```python
constante = tf.constant(15)
matrix1 = tf.fill((6,6),10)
matrix2 = tf.random_normal((5,5))
matrix3 = tf.random_uniform((4,4), minval=0, maxval=5)
matrix_cero = tf.zeros((2,2))
matrix_ones = tf.ones((3,3))
```

```python
operaciones = [constante, matrix1, matrix2, matrix3, matrix_cero, matrix_ones]
```

```python
with tf.Session() as sesion:
    for op in operaciones:
        print(sesion.run(op))
        print('\n')
```
